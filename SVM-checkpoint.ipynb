```python
import os

import numpy as np
import pandas as pd 

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from skimage.io import imread
from skimage.color import rgb2gray

```

```python
from skimage.io import imshow
img = imread("smile\James_Jones_0001.jpg")
```

```python
img.shape
rgb2gray(img).shape
```

```python
plt.imshow(rgb2gray(img),cmap="gray")
```

```python
smile_path = "smile"
nsmile_path = "non_smile"
smile_file = os.listdir(smile_path)
nsmile_file = os.listdir(nsmile_path)

print("# smile files {},\n # non smile files {}".format(len(smile_file),
                                                       len(nsmile_file)))
```

```python
x_smile = np.empty((len(smile_file),64*64))
x_nsmile = np.empty((len(smile_file),64*64))

y = []
```

```python
x_smile[1,:].shape
```

```python
for i,f in enumerate(smile_file):
    img = rgb2gray(imread(os.path.join(smile_path,f)))
    x_smile[i,:] = np.reshape(img,(64*64))
    y.append(1)
    
```

```python
for i,f in enumerate(smile_file):
    img = rgb2gray(imread(os.path.join(smile_path,f)))
    x_smile[i,:] = np.reshape(img,(64*64))
    y.append(0)
```

```python
x = np.concatenate([x_smile,x_nsmile],axis=0)
y = np.array(y)
```

```python
x_smile[1,:].shape
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.decomposition import PCA
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)

print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
clf = Pipeline(steps=[("dimention reduction",PCA(n_components=500)),
                     ("classifier",SVC(kernel="linear",C=10000))])
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

```python
y_pred = clf.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_estimator(clf,x_test,y_test)
```

```python
print("SVM CLASSIFICATION REPORT: ")
print(classification_report(y_true=y_test,y_pred=y_pred))
```
